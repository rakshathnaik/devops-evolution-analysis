```python
import matplotlib.pyplot as plt
import pandas as pd

# Load and Clean
df = pd.read_csv('DevOps efficiency.csv')
df['Lead_Time_Numeric'] = df['Lead Time (Hours)'].astype(str).str.extract(r'(\d+\.?\d*)').astype(float)

# Plotting
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 15))

# Panel 1: Frequency
ax1.bar(df['Quarter'], df['Deploys per Week'], color='skyblue', edgecolor='black')
ax1.set_title('1. Deployment Frequency', fontsize=14, fontweight='bold')
ax1.set_ylabel('Deploys / Week')

# Panel 2: Lead Time
ax2.plot(df['Quarter'], df['Lead_Time_Numeric'], color='red', marker='o', linewidth=3)
ax2.set_title('2. Lead Time for Changes', fontsize=14, fontweight='bold')
ax2.set_ylabel('Hours')

# Panel 3: Failure Rate
ax3.bar(df['Quarter'], df['Change Failure Rate (%)'], color='lightgreen', edgecolor='black')
ax3.set_title('3. Change Failure Rate', fontsize=14, fontweight='bold')
ax3.set_ylabel('Failure %')

plt.tight_layout()
plt.savefig('detailed_devops_metrics.png', dpi=300)
plt.show()
```
